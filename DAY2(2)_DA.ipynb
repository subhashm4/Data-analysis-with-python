```python
sub=['psychology','nanology','medicine','oncology','anthropology']
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#student 1
plt.plot(sub,s1,label="student 1",color="k")
#student 2
plt.plot(sub,s2,label="student 2",color="b")
#cutoffm
plt.plot(sub,cut,label="cutoff",color="r")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
sub=['psychology','nanology','medicine','oncology','anthropology']
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#student 1
plt.scatter(sub,s1,label="student 1",color="k",marker="p")
#student 2
plt.scatter(sub,s2,label="student 2",color="b",marker="p")
#cutoff
plt.plot(sub,cut,label="cutoff",color="r",linewidth="5",linestyle=":")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
#an india vs australia match is on live india has even 298 target for australia,the runs for every over is given as follows [25,51,84,131,160,189,220,250,267,297]
#Australia have completed 25 overs 15,41,94,110,151,
#plot overs vs runs for boththe countries with india has a blue line and australia has a yellow line with required labels and legend
#the marker for india is flag and marker for australia is plus
#take overs from a range functions create runs as arrays
import numpy as np
from matplotlib import pyplot as plt
A=np.array([25,51,84,131,160,189,220,250,267,297])
B=np.array([15,41,94,110,151])
o=np.arange(5,51,5)
o1=np.arange(5,26,5)
plt.xlabel("Runs")
plt.ylabel("Score")
plt.plot(o,A,label="india",color='b',marker='>',linewidth=2)
plt.plot(o1,B,label="australia",color='y',marker='+',linewidth=2)
plt.title("INDIA VS AUSTRALIA")
plt.grid()
plt.legend()
```
