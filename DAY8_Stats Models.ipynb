```python
#hypothesis testing:
import numpy as np
from scipy import stats
```

```python
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
t_stat, p_value = stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
#Set significance level
alpha = 0.05
```

```python
if p_value < alpha:
  print("Reject the null hypothesis.The means are significant")
else:
  print("Fail to reject null hypothesis.The means are not signoficant")
```

```python
import pandas as pd
df = pd.read_csv('/content/Vaccine.csv')
df
```

```python
a=df['Efficiency'].mean()
b=df['Rec_Rate'].mean()
print("Mean Recovery",b)
group_A =df['Efficiency']
group_B =df['Rec_Rate']
t_stat, p_value = stats.ttest_ind(group_A,group_B)
print("t-val",t_stat)
print("p-val",p_value)
```

```python
alpha = 0.05
```

```python
if p_value >alpha:
  print("H0-related")
else:
  print("H1- not related")
```

# ANOVA(one way ANOVA)

```python
dfA = df[df['Vaccine']=='A']['Efficiency']
dfB = df[df["Vaccine"]=='B']['Efficiency']
dfA
```

```python
f_stat, p_value = stats.f_oneway(dfA,dfB)

print("F-statistics:", f_stat)
print("p-value:", p_value)
```
