```python
#list comprehension
#new list=[output for if]
a=[23,58,69,47,15,62]
b=[i*i for i in a if i>50]
print(b)
```

```python
#lc
#new list=[(statement of if) if (condition) else (statement of else) for]
l=[23,67,12,34,69]
res=[i*i if i>50 else 'X' for i in l]
print(res)
```

```python
#set operations
s1={1,2,3,4,5}
s2={4,5,6,7,8}
#union
print(s1.union(s2))
#intersection
print(s1.intersection(s2))
#difference
print(s1.difference(s2))
#symmetric_difference
print(s1.symmetric_difference(s2))
#subset
print(s1.issubset(s2))
#superset
print(s1.issuperset(s2))
#disjoint
print(s1.isdisjoint(s2))
```
