```python
from bs4 import BeautifulSoup
import requests as r
```

```python
ar="https://aniwatch.to/tv"
```

```python
page=r.get(ar)
```

```python
code=BeautifulSoup(page.text,'html')
print(code)
```

```python
#accessing single tag or single element
t=code.find("title")
print(t)
```

```python
print(t.text)
```

```python
arr="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
page=r.get("https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1")
code=BeautifulSoup(page.text,'html')
print(code)
```

```python
t=code.find("table")
print(t)
```

```python
h=code.find_all("th")
print(h)
```

```python
headings=[i.text for i in h]
print(headings)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=headings)
print(mydf)
```

```python
rows=code.find_all('tr')
print(rows)
```

```python
td=code.find_all('td')
td
```

```python
data_act=[i.text for i in td]
print(data_act)
```

```python
ind=0
st=0
stop=4
for i in data_act:
  if stop <=40:
    mydf.loc[ind]=data_act[st:stop]
    st=st+4
    stop=stop+4
    ind=ind+1
print(mydf)
```

```python
mydf.to_csv("rich.csv")
```
